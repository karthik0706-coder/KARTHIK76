```python

import pandas as pd
from sklearn.linear_model import LinearRegression

# Sample training data
data = {
    'room_size': [100, 200, 300, 400, 500],
    'material_quality': [1, 2, 2, 3, 3],
    'design_complexity': [1, 2, 2, 3, 3],
    'estimated_cost': [5000, 12000, 15000, 25000, 30000]
}

df = pd.DataFrame(data)
X = df[['room_size', 'material_quality', 'design_complexity']]
y = df['estimated_cost']

# Train model
model = LinearRegression()
model.fit(X, y)

# Chatbot logic
def chatbot():
    print("Hello! I am your Interior Cost Estimator Bot.")
    print("I'll ask you a few questions to estimate your interior project cost.\n")

    room_size = float(input("1. What is the room size in sq ft? "))

    print("\n2. What is the material quality?")
    print("   1 = Standard")
    print("   2 = Premium")
    print("   3 = Luxury")
    material_quality = int(input("   Enter your choice (1/2/3): "))

    print("\n3. What is the design complexity?")
    print("   1 = Simple")
    print("   2 = Moderate")
    print("   3 = Complex")
    design_complexity = int(input("   Enter your choice (1/2/3): "))

    # Create input dataframe
    user_df = pd.DataFrame([{
        'room_size': room_size,
        'material_quality': material_quality,
        'design_complexity': design_complexity
    }])

    # Predict cost
    predicted_cost = model.predict(user_df)
    print(f"\nEstimated Interior Cost: ${predicted_cost[0]:,.2f}")
    print("Thank you for using the bot!")

# Run chatbot
chatbot()
```
