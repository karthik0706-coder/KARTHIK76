```python
import os

import PyPDF2

from langchain.prompts import PromptTemplate

from langchain_google_genai import ChatGoogleGenerativeAI


# ======== SETUP ========

os.environ["GOOGLE_API_KEY"] = "AIzaSyBsffBWbmOyQVnJ58I3lQZ02RuhSWPByqA"  # Replace with your Gemini 2.5 API key


# Load Gemini 2.5 Pro via LangChain

llm = ChatGoogleGenerativeAI(

    model="gemini-pro",  # Gemini 2.5 Pro uses this endpoint

    temperature=0.5,

    max_output_tokens=2048

)


# ======== PDF TEXT EXTRACTION ========

def extract_text_from_pdf(pdf_path):

    text = ""

    with open(pdf_path, "rb") as file:

        reader = PyPDF2.PdfReader(file)

        for page in reader.pages:

            text += page.extract_text()

    return text


# ======== PROMPT TEMPLATE ========

prompt_template = PromptTemplate(

    input_variables=["resume_text"],

    template="""

You are an expert career consultant. Analyze the following resume and give detailed, professional feedback:

- Identify strengths

- Highlight missing or weak sections

- Suggest improvements for job readiness

- Recommend formatting and wording enhancements


Resume Text:

{resume_text}


Your Feedback:

"""

)


# ======== ANALYSIS FUNCTION ========

def analyze_resume_with_gemini(pdf_path):

    resume_text = extract_text_from_pdf(pdf_path)

    filled_prompt = prompt_template.format(resume_text=resume_text)

    response = llm.invoke(filled_prompt)

    return response.content


# ======== RUN ========

if __name__ == "__main__":

    pdf_path = "/mnt/data/file-JLYoDCfKAHANuitGCxxZgm"  # Your uploaded resume

    feedback = analyze_resume_with_gemini(pdf_path)

    print("\n===== Resume Feedback from Gemini 2.5 Pro =====\n")

    print(feedback)

``
```
