```python

from PIL import Image, ImageDraw, ImageFont

def create_image(output_path="sample_image.png"):
    # Create a blank image with RGB mode and white background
    width, height = 400, 300
    image = Image.new("RGB", (width, height), "white")

    # Draw on the image
    draw = ImageDraw.Draw(image)
    draw.rectangle([(50, 50), (350, 250)], outline="blue", width=5)
    draw.ellipse([(150, 100), (250, 200)], fill="red", outline="black")

    # Optional: Add text
    try:
        font = ImageFont.truetype("arial.ttf", 20)
    except IOError:
        font = ImageFont.load_default()

    draw.text((120, 20), "Hello, Pillow!", fill="black", font=font)

    # Save the image
    image.save(output_path)
    print(f"Image saved to {output_path}")

# Generate the image
create_image()
```
